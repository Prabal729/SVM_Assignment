 1. What is a Support Vector Machine (SVM) ?
 -> An SVM is a supervised machine learning algorithm used for classification and regression tasks. It finds an optimal hyperplane that best separates data points into different classes.
 2. What is the difference between Hard Margin and Soft Margin SVM ?
 -> Hard Margin SVM: Seeks a hyperplane that perfectly separates all data points without any misclassifications. It's sensitive to outliers and only works for linearly separable data.
  Soft Margin SVM: Allows for some misclassifications or data points to fall within the margin, making it more robust to noise and applicable to non-linearly separable data.
 3. What is the mathematical intuition behind SVM ?
 -> SVM aims to maximize the margin (the distance between the hyperplane and the closest data points of each class) while minimizing classification errors. It's essentially an optimization problem.
 4. What is the role of Lagrange Multipliers in SVM ?
 -> Lagrange Multipliers are used to solve the constrained optimization problem of SVM, allowing us to find the optimal hyperplane while satisfying the conditions that data points are on the correct side of the margin.
 5. What are Support Vectors in SVM ?
 -> Support Vectors are the data points closest to the separating hyperplane. They are the most influential points in defining the hyperplane and the margin.
 6. What is a Support Vector Classifier (SVC) ?
 -> SVC is the classification variant of SVM, used for classifying data points into distinct categories.
 7. What is a Support Vector Regressor (SVR) ?
 -> SVR is the regression variant of SVM, used for predicting continuous output values.
 8. What is the Kernel Trick in SVM ?
 -> The Kernel Trick allows SVM to implicitly map data into a higher-dimensional feature space without explicitly calculating the coordinates in that space. This enables SVM to find non-linear decision boundaries in the original feature space.
 9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel ?
 -> Linear Kernel: Suitable for linearly separable data, creating a straight decision boundary.
Polynomial Kernel: Creates non-linear decision boundaries by considering polynomial combinations of features.
RBF (Radial Basis Function) Kernel: Creates highly non-linear and complex decision boundaries, effective for non-linearly separable data.
 10. What is the effect of the C parameter in SVM ?
 -> The C parameter controls the trade-off between maximizing the margin and minimizing misclassifications.

Small C: Prioritizes a larger margin, allowing more misclassifications (underfitting).
Large C: Prioritizes fewer misclassifications, leading to a smaller margin (potential overfitting).
 11. What is the role of the Gamma parameter in RBF Kernel SVM ?
 -> The gamma parameter defines the influence of a single training example in the RBF kernel.

Small gamma: Larger influence, resulting in a smoother decision boundary (potential underfitting).
Large gamma: Smaller influence, resulting in a more complex, wiggly decision boundary (potential overfitting).

 12. What is the Naïve Bayes classifier, and why is it called "Naïve" ?
 -> Naïve Bayes is a probabilistic classifier based on Bayes' Theorem. It's called "Naïve" because it assumes strong (naïve) independence between features, meaning the presence of one feature does not affect the presence of any other feature.

 13. What is Bayes’ Theorem ?
 -> Bayes' Theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event.
 14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes ?
 -> Gaussian Naïve Bayes: Assumes features follow a Gaussian (normal) distribution. Used for continuous data.
Multinomial Naïve Bayes: Assumes features represent counts (e.g., word counts in text). Used for discrete data like text classification.
Bernoulli Naïve Bayes: Assumes binary features (e.g., presence or absence of a word). Used for binary data.
 15. When should you use Gaussian Naïve Bayes over other variants ?
 -> Use Gaussian Naïve Bayes when your features are continuous and can be reasonably assumed to follow a normal distribution.
 16. What are the key assumptions made by Naïve Bayes ?
 -> The key assumption is conditional independence of features given the class label.
 17. What are the advantages and disadvantages of Naïve Bayes ?
 -> Advantages: Simple, fast to train, works well with high-dimensional data, good for text classification, robust to irrelevant features.
Disadvantages: Strong independence assumption can be violated in real-world data, struggles with correlated features, probability estimates can be inaccurate.
 18. Why is Naïve Bayes a good choice for text classification ?
 -> Naïve Bayes is good for text classification due to its simplicity, speed, and effectiveness with high-dimensional sparse data (like word counts) even with the independence assumption.
 19. Compare SVM and Naïve Bayes for classification tasks ?
 -> SVM: Focuses on finding an optimal separating hyperplane, often performs well with complex, non-linear relationships, less sensitive to feature independence. Can be slower to train on very large datasets.
Naïve Bayes: Probabilistic, fast, scales well to large datasets and high dimensions, but relies on a strong independence assumption that might not hold true.
 20. How does Laplace Smoothing help in Naïve Bayes?
 -> Laplace Smoothing (add-one smoothing) addresses the "zero-frequency problem." If a word or feature doesn't appear in the training data for a particular class, its probability would be zero, making the entire posterior probability zero. Laplace Smoothing adds a small count (typically 1) to all feature counts, preventing zero probabilities and making the model more robust.

1.  Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC()

svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM Classifier on Iris dataset: {accuracy:.2f}")
```

4. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

X, y = make_classification(
    n_samples=200,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    class_sep=0.8,
    random_state=42
)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

svm_clf = SVC(kernel='poly', degree=3, C=1.0)
svm_clf.fit(X_scaled, y)

def plot_decision_boundary(model, X, y):
    h = 0.02  # Mesh step size
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title("SVM with Polynomial Kernel (degree=3)")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid(True)
    plt.show()

plot_decision_boundary(svm_clf, X_scaled, y)
```

6. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20

10. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and
after Laplace Smoothing

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score
from sklearn.datasets import fetch_openml

data = fetch_openml("mushroom", version=1, as_frame=True)
df = data.frame

for col in df.columns:
    df[col] = df[col].astype('category').cat.codes

df.replace(-1, 0, inplace=True)

X = df.drop('class', axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model_no_smoothing = CategoricalNB(alpha=0)
model_no_smoothing.fit(X_train, y_train)
y_pred_no_smoothing = model_no_smoothing.predict(X_test)

model_laplace = CategoricalNB(alpha=1)
model_laplace.fit(X_train, y_train)
y_pred_laplace = model_laplace.predict(X_test)

acc_no_smoothing = accuracy_score(y_test, y_pred_no_smoothing)
acc_laplace = accuracy_score(y_test, y_pred_laplace)

print("Accuracy without Laplace Smoothing (alpha=0):", acc_no_smoothing)
print("Accuracy with Laplace Smoothing (alpha=1):", acc_laplace)
```

15.  Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare
results

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model_all_features = GaussianNB()
model_all_features.fit(X_train, y_train)
y_pred_all = model_all_features.predict(X_test)
acc_all = accuracy_score(y_test, y_pred_all)

selector = SelectKBest(score_func=chi2, k=2)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

model_selected = GaussianNB()
model_selected.fit(X_train_selected, y_train)
y_pred_selected = model_selected.predict(X_test_selected)
acc_selected = accuracy_score(y_test, y_pred_selected)

print("Accuracy with all features       :", acc_all)
print("Accuracy with selected features  :", acc_selected)

selected_features = selector.get_support(indices=True)
print("Selected feature indices:", selected_features)
print("Selected feature names  :", [data.feature_names[i] for i in selected_features])
```

16.  Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO)
strategies on the iris dataset and compare their accurac

```python

from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier

iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target

X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

svm_base = SVC(kernel='linear')

ovr_classifier = OneVsRestClassifier(svm_base)
ovr_classifier.fit(X_train_iris, y_train_iris)
y_pred_ovr = ovr_classifier.predict(X_test_iris)
accuracy_ovr = accuracy_score(y_test_iris, y_pred_ovr)
print(f"Accuracy of SVM with One-vs-Rest strategy: {accuracy_ovr:.2f}")

ovo_classifier = OneVsOneClassifier(svm_base)
ovo_classifier.fit(X_train_iris, y_train_iris)
y_pred_ovo = ovo_classifier.predict(X_test_iris)
accuracy_ovo = accuracy_score(y_test_iris, y_pred_ovo)
print(f"Accuracy of SVM with One-vs-One strategy: {accuracy_ovo:.2f}")

print("\nStrategy Comparison:")
print(f"One-vs-Rest Accuracy: {accuracy_ovr:.2f}")
print(f"One-vs-One Accuracy:  {accuracy_ovo:.2f}")
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

iris = datasets.load_iris()
X = iris.data

n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

accuracy_scores = []

base_svm = SVC(kernel='rbf')

for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    print(f"Fold {fold+1}/{n_splits}")


    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]


    svm_classifier = clone(base_svm)


    svm_classifier.fit(X_train, y_train)


    y_pred = svm_classifier.predict(X_test)


    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    print(f"  Accuracy: {accuracy:.4f}")

average_accuracy = np.mean(accuracy_scores)

print(f"\nAverage Accuracy using Stratified {n_splits}-Fold Cross-Validation: {average_accuracy:.4f}")
```

19. Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare
performance.

```python
data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

unique_classes, class_counts = np.unique(y_train, return_counts=True)
empirical_priors = class_counts / len(y_train)

print("Empirical Priors:", empirical_priors)

gnb_default = GaussianNB()
gnb_default.fit(X_train, y_train)
y_pred_default = gnb_default.predict(X_test)
acc_default = accuracy_score(y_test, y_pred_default)
print("\nAccuracy with Default Priors:", acc_default)

gnb_empirical = GaussianNB(priors=empirical_priors)
gnb_empirical.fit(X_train, y_train)
y_pred_empirical = gnb_empirical.predict(X_test)
acc_empirical = accuracy_score(y_test, y_pred_empirical)
print("Accuracy with Empirical Priors:", acc_empirical)

num_classes = len(np.unique(y_train))
equal_priors = [1/num_classes] * num_classes
print("Equal Priors:", equal_priors)

gnb_equal = GaussianNB(priors=equal_priors)
gnb_equal.fit(X_train, y_train)
y_pred_equal = gnb_equal.predict(X_test)
acc_equal = accuracy_score(y_test, y_pred_equal)
print("Accuracy with Equal Priors:", acc_equal)

skewed_priors = [0.8, 0.1, 0.1]
print("Skewed Priors:", skewed_priors)

try:
    gnb_skewed = GaussianNB(priors=skewed_priors)
    gnb_skewed.fit(X_train, y_train)
    y_pred_skewed = gnb_skewed.predict(X_test)
    acc_skewed = accuracy_score(y_test, y_pred_skewed)
    print("Accuracy with Skewed Priors:", acc_skewed)
except ValueError as e:
    print(f"Could not use skewed priors: {e}")
    print("Check if the sum of priors equals 1 and the number of priors matches the number of classes.")

print("\nPerformance Comparison with Different Priors:")
print(f"Default Priors Accuracy:   {acc_default:.4f}")
print(f"Empirical Priors Accuracy: {acc_empirical:.4f}")
print(f"Equal Priors Accuracy:     {acc_equal:.4f}")
if 'acc_skewed' in locals():
     print(f"Skewed Priors Accuracy:    {acc_skewed:.4f}")
```

20. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy

```python
from sklearn.feature_selection import RFE

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_no_rfe = SVC(kernel="linear")
svm_no_rfe.fit(X_train, y_train)
y_pred_no_rfe = svm_no_rfe.predict(X_test)
accuracy_no_rfe = accuracy_score(y_test, y_pred_no_rfe)
print(f"Accuracy of SVM without RFE: {accuracy_no_rfe:.4f}")

estimator = SVC(kernel="linear")

selector = RFE(estimator, n_features_to_select=2, step=1)
selector = selector.fit(X_train, y_train)

X_train_rfe = selector.transform(X_train)
X_test_rfe = selector.transform(X_test)

svm_with_rfe = SVC(kernel="linear")
svm_with_rfe.fit(X_train_rfe, y_train)
y_pred_with_rfe = svm_with_rfe.predict(X_test_rfe)
accuracy_with_rfe = accuracy_score(y_test, y_pred_with_rfe)
print(f"Accuracy of SVM with RFE: {accuracy_with_rfe:.4f}")

print("\nSelected features indices (0-based):", selector.support_)
print("Ranking of features (1 is best):", selector.ranking_)


print("\nAccuracy Comparison:")
print(f"Accuracy without RFE: {accuracy_no_rfe:.4f}")
print(f"Accuracy with RFE:    {accuracy_with_rfe:.4f}")
```

25. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC
score

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import LabelEncoder

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

y_binary = (y == 2).astype(int)

X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(
    X, y_binary, test_size=0.3, random_state=42, stratify=y_binary
)

gnb_classifier = GaussianNB()
gnb_classifier.fit(X_train_binary, y_train_binary)

y_pred_proba = gnb_classifier.predict_proba(X_test_binary)[:, 1]

roc_auc = roc_auc_score(y_test_binary, y_pred_proba)

print(f"ROC-AUC Score of Gaussian Naïve Bayes Classifier (Binary Problem): {roc_auc:.4f}")

fpr, tpr, thresholds = roc_curve(y_test_binary, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
